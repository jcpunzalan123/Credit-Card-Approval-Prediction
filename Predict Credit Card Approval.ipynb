## Feature Engineering

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
OE = OneHotEncoder()
LE =LabelEncoder()
```

```python
data = data.drop(columns=['Ind_id', 'Mobile_phone', 'Work_Phone', 'Phone', 'Email_id'])
```

```python
features = ['Type_Income', 'Education', 'Marital_status', 'Housing_type', 'Type_Occupation']

features_oe = OE.fit_transform(data[features])
encoded_features = pd.DataFrame(features_oe.toarray(), columns=OE.get_feature_names_out(features), dtype=np.int8)
data = pd.concat([data.reset_index(drop=True), encoded_features.reset_index(drop=True)], axis=1)
data.drop(features, axis=1, inplace=True)
data.head()
```

```python
for col in data:
    if data[col].dtype == 'object':
        data[col] = LE.fit_transform(data[col])
```

```python
data.insert(len(data.columns)-1, 'label', data.pop('label'))
```

## Model Building

```python
from sklearn.model_selection import train_test_split
# spliting the data 
x = data.iloc[:,0:-1]
y = data.iloc[:,-1]
```

```python
y
```

### Split the train and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#Check the dimensions of train and test data
print(f'Shape train: {x_train.shape} \nShape test: {x_test.shape}')
```

### Scale the data using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))

rescale_x_train = scaler.fit_transform(x_train)
rescale_x_test = scaler.fit_transform(x_test)
```

```python
#Check the dimensions of train and test data
print(f'Shape train: {rescale_x_train.shape} \nShape test: {rescale_x_test.shape}')
```

### Implement Logistic Regression

```python
# importing the LogisticRegression
from sklearn.linear_model import LogisticRegression
logreg_model = LogisticRegression(solver='lbfgs', max_iter=3000, random_state=0)
logreg_model.fit(rescale_x_train, y_train)
```

```python
# Evaluate Logistic Regression Model
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred = logreg_model.predict(rescale_x_test)

print("Accuracy of Logistic Regression classifier: ", accuracy_score(y_test, y_pred))

print(confusion_matrix(y_test,y_pred))
```

### Implement Decision Tree

```python
# importing the LogisticRegression
from sklearn.tree import DecisionTreeClassifier
Dtree_model = DecisionTreeClassifier()
Dtree_model.fit(rescale_x_train, y_train)
```

```python
# Evaluate Decision Tree Model
y_pred = Dtree_model.predict(rescale_x_test)

print("Accuracy of Decision Tree classifier: ", accuracy_score(y_test, y_pred))

print(confusion_matrix(y_test,y_pred))
```

```python
# cross validation
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

# model= DecisionTreeClassifier()
ssplit=ShuffleSplit(n_splits=10,test_size=0.20)

results=cross_val_score(Dtree_model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```

### Implement XG Boost

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

XGB_model = XGBClassifier()
XGB_model.fit(x_train,y_train)
```

```python
# Evaluate XGBoost Model
y_pred = XGB_model.predict(x_test)

print("Accuracy of XGBoost classifier: ", accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# cross validation
ssplit = ShuffleSplit(n_splits=10,test_size=0.20)
results=cross_val_score(XGB_model,x,y,cv=ssplit)

print(results)
print(np.mean(results))
```

### Implement Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

RF_model = RandomForestClassifier()
RF_model.fit(x_train,y_train)
```

```python
# Evaluate Random Forest Model
y_pred = RF_model.predict(x_test)

print("Accuracy of Random Forest classifier: ", RF_model.score(x_test, y_test))
# Evaluate the confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# cross validation
ssplit = ShuffleSplit(n_splits=10,test_size=0.20)

results = cross_val_score(RF_model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```

## Conclusion

After building multiple classification models, we can compare their respective accuracy:

- Logistic Regression: 90.93%
- Decision Tree: 77.19%
- XGBoost: 92.01%
- Random Forest: 93.74%

Based from the results, Random Forest classifier has the highest accuracy.
